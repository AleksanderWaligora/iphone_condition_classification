```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
driver =webdriver.Chrome()
```

```python
driver.get("https://www.ebay.pl/sch/i.html?_from=R40&_nkw=iphone&_sacat=0&_ipg=240&_dmd=2&rt=nc&_pgn=1")	
```

```python
driver.find_element(By.ID,"gdpr-banner-accept").click()
```

```python
import pandas as pd
nameList=[]
sellerList=[]
conditionList=[]
priceList=[]
shipping_costList=[]
shipping_countryList=[]
free_returnList=[]
sponsoredList=[]
badgeList=[]
buyNowList=[]

for i in range(1,5):
    driver.get(f"https://www.ebay.pl/sch/i.html?_from=R40&_nkw=iphone&_sacat=0&_ipg=240&_dmd=2&rt=nc&_pgn={i}")	
    parentElement=driver.find_element(By.ID,"srp-river-results")
    elementList = parentElement.find_elements(By.XPATH,"//li[@class='s-item s-item__dsa-on-bottom s-item__pl-on-bottom']")
    for element in elementList:
        name=None
        seller=None
        condition=None
        price=None
        shipping_cost=None
        shipping_country=None
        free_return=None
        sponsored=None
        badge=None
        buyNow=None
        
        name =element.find_element(By.CLASS_NAME,"s-item__title").text
        nameList.append(name)
       
        stan_seller = element.find_element(By.CLASS_NAME,"s-item__subtitle").text
        
        condition = stan_seller.split("|")[0]
        conditionList.append(condition)
       
        seller = stan_seller.split("|")[1].replace("Sprzedawca:","")
        sellerList.append(seller)
       
        price=element.find_element(By.CLASS_NAME,"s-item__price").text.replace("zł","")
        priceList.append(price)
    
        try:
            shipping_cost=element.find_element(By.XPATH,".//span[@class='s-item__shipping s-item__logisticsCost']").text.replace("zł za wysyłkę","").replace("+","").replace("(szac.)","").strip()
            
        except Exception:
            shipping_cost="0"
        shipping_costList.append(shipping_cost)
        
        shipping_country = element.find_element(By.XPATH,".//span[@class='s-item__location s-item__itemLocation']").text.replace("z:","")
        shipping_countryList.append(shipping_country)
    
        try:
            free_return = element.find_element(By.XPATH,".//span[@class='s-item__free-returns s-item__freeReturnsNoFee']").text
            free_return=True
        except Exception:
            free_return=False
    
        free_returnList.append(free_return)
    
        try:
            sponsored= element.find_element(By.XPATH,".//span[@class='s-item__dynamic s-item__lastOne']")
            sponsored=True
        except Exception:
            sponsored=False
        sponsoredList.append(sponsored)
    
        try:
            badge = element.find_element(By.XPATH,".//span[@class='s-item__etrs-badge clipped']").text
            badge=True
        except Exception:
            badge=False
        badgeList.append(badge)
    
        try:
            buyNow = element.find_element(By.XPATH,".//span[@class='s-item__purchase-options s-item__purchaseOptions']")
            buyNow=True
        except Exception:
            buyNow=False
        buyNowList.append(sponsored)
        print(name)
        
```

## Tworzenie data frame

```python

import numpy as np
df = pd.DataFrame(columns=["Name","Seller","Condition","Price","Shipping Cost","Shipping Country","Free Return","Last Item","Best Seller","Buy Now"])
df["Name"]=np.array(nameList)
df["Seller"]=np.array(sellerList)
df["Condition"]=np.array(conditionList)
df["Price"]=np.array(priceList)
df["Shipping Cost"]=np.array(shipping_costList)
df["Shipping Country"]=np.array(shipping_countryList)
df["Free Return"]=np.array(free_returnList)
df["Last Item"]=np.array(sponsoredList)
df["Best Seller"]=np.array(badgeList)
df["Buy Now"]=np.array(buyNowList)
df
```

## Stworzenie kategori Generation

```python
def extract_iphone_model(name):
    name=name.lower()
    name_list = name.split(" ")
    gen=None
    try:
        start_index = name_list.index("iphone")
        gen=name_list[start_index+1]
    except Exception:
            gen=None
    
    return gen
   

df["Generation"]=df["Name"].apply(extract_iphone_model)
df["Generation"] = df["Generation"].astype("category")
df
```

```python
df.to_csv("ebay_iphone_uncleaned.csv",index=False)
```
